```python
import data
import random
from sklearn.decomposition import PCA
```

### Loading Dataset

```python
Data_vecs ,Data_labels = data.load_data("/content/drive/MyDrive/251B/PA1" , train = True)
# Data_vecs = np.array(list(Data_vecs))
# Data_labels = np.array(list(Data_labels))
# X,t = data.shuffle((Data_vecs,Data_labels))
```

### Main PCA

```python
def PCAMine(X_meanshifted,p):
  pca = PCA(n_components = p)
  pca.fit(X_meanshifted) # Do SVD and get vectors
  return pca
```

### 3.1) Choosing Random image and performing PCA

```python
fig = plt.figure(figsize=(10,5))

PC = [2,10,50,100,200,784]
randi = random.randint(0,len(Data_vecs)) # Random Sample

# Orig Image plot
fig.add_subplot("24"+str(1))
plt.imshow(Data_vecs[randi].reshape([28,28]),cmap="gray")
plt.title("Original Image")
c=2

mean = np.mean(Data_vecs,axis = 0)  # Removing Mean
X =Data_vecs - mean

print("Random Index: ", randi)
# All PCs
for p in PC:
  pca, mean = PCAMine(X,p) # fit PC
  pcX = pca.transform(X) # PCA
  invX = pca.inverse_transform(pcX) + mean # inverse PCA

  # Plotting
  fig.add_subplot("24"+str(c))
  plt.title("PCA #Comp "+str(p))
  plt.imshow(invX[randi].reshape([28,28]),cmap="gray")
  c+=1

plt.show()
```

### 3.2) Visualizing weights of 2 classes

```python
import data
def vizn1n2(n1,n2):
  Data_vecs , Data_labels = data.load_data("/content/drive/MyDrive/251B/PA1" , train = True)
  Data_vecs = np.array(list(Data_vecs))
  Data_labels = np.array(list(Data_labels))
  Data_vecs_normalized = data.min_max_normalize(Data_vecs)
  # Data_vecs_normalized = data.append_bias(Data_vecs_normalized)

  X8, t8 = data.generate_nDS(Data_vecs_normalized, Data_labels,n1)
  X5, t5 = data.generate_nDS(Data_vecs_normalized, Data_labels,n2)

  t8[:] = 0
  t5[:] = 1

  X = np.vstack((X8,X5))
  t = np.hstack((t8,t5))

  X,t = data.shuffle((X,t))

  #####

  fig = plt.figure(figsize=(16,8))

  p = 10
  mean = np.mean(X,axis = 0)  # Removing Mean
  X =X - mean
  pca = PCAMine(X,p) # fit PC
  c=0

  for i in pca.components_:
    fig.add_subplot("25"+str(c))
    plt.title("PCA Component #"+str(c))
    plt.imshow(i.reshape([28,28]),cmap="gray")
    c+=1

  plt.show()
```

```python
import data
X , y = data.load_data("/content/drive/MyDrive/251B/PA1" , train = True)
Xtest , ytest = data.load_data("/content/drive/MyDrive/251B/PA1" , train = False)
```

## Random 10 classes

```python
def plot_mnist_sample_high_res(X_train, y_train):
    """
    This function plots a sample image for each category,
    The result is a figure with 2x5 grid of images.
    
    """
    plt.figure()
    
    """ ==========
    YOUR CODE HERE
    ========== """
    seti = set()
    c = 0
    imgs = [[0]]*10
    X = X_train
    y = y_train
    while len(seti)!=10:
      categ = y[c]
      if categ not in seti:
        # print(len(seti))
        imgs[categ] = X[c]
        seti.add(categ)
      
      c+=1

    # print(imgs)
    fig = plt.figure(figsize=(12, 9))

    for i in range(5):
      tmp = "25"+str(i+1)
      ax1 = fig.add_subplot(int(tmp))
      plt.imshow(imgs[i].reshape(28,28), cmap="gray")
      plt.title("Number: "+str(i))


    for i in range(5):
      ax1 = fig.add_subplot(2,5,i+1+5)
      plt.imshow(imgs[i+5].reshape(28,28), cmap="gray")
      plt.title("Number: "+str(i+5))

  
    plt.show()
```

```python
import matplotlib.pyplot as plt
X , y = data.shuffle((X,y))
plot_mnist_sample_high_res(X,y)
```

```python
import numpy as np

# Create a sample matrix
matrix = np.array([[1, 2, 3, 0], [4, 5, 6, 0], [7, 8, 9, 0]])

# Initialize a variable to store the average of non-zero elements per column
average_per_column = np.zeros(matrix.shape[1])

# Iterate through each column
for i in range(matrix.shape[1]):
    # Get the column
    column = matrix[:, i]
    # Removing 0 from the column
    column = column[column != 0]
    if column.size != 0:
        # Calculate the average of the non-zero elements in the column
        average_per_column[i] = np.mean(column)
    else:
        average_per_column[i] = 0

# Replace the NaN values in the array with 0
average_per_column = np.nan_to_num(average_per_column)

print(average_per_column)
```
